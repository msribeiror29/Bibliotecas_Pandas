## Junção das Bases de Dados

Realizando um join utilizando a tabela client


Agora que entendemos a estrutura das duas bases, podemos juntá-las para enriquecer a análise.


```python
client = pd.read_excel('/content/cliente_varejo.xlsx')
client.head()
```

```python
salles_right.head()
```

Observando as duas tabelas vemos que ha conexao entre elas através da coluna cliente_log onde faremos o join.

Analisando a tabela client

```python
client[['idade', 'renda']].describe()
```

```python
client.info()
```

```python
client = client.astype({'renda':float})
client.info()
```

```python
salles_client = salles_right.merge(client, how='left', on= 'cliente_Log')
salles_client.head()
```

Desta forma, temos uma base de dados unificada com informações tanto de vendas quanto de clientes.

## Cálculo das Métricas

Com os dados tratados e unificados, podemos partir para a análise e cálculo das métricas solicitadas:

- Média de vendas por mês
- Percentual de vendas por canal
- Média de renda por canal
- Média de idade por bandeira

Vamos utilizar groupby, agregações e manipulações simples do Pandas:


Média de renda para cada tipo de canal de venda

```python
agg_id_income = round( salles_client.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_id_income
```

Média de idade de clientes por bandeira.

```python
agg_id_bandeira = round(salles_client.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_id_bandeira
```

## Visualização de Dados

Por fim, para facilitar a apresentação dos resultados, podemos construir visualizações e gráficos utilizando bibliotecas como Matplotlib e Seaborn.


```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (7,4))
plt.bar (agg_id_bandeira ['bandeira'], agg_id_bandeira['idade'], color = 'blue')
plt.ylabel ('Média de Idade')
plt.title ('Idade Média por Idade', loc ='center', fontsize = '18', fontweight = 'bold', color = 'black');
```

```python
fig = plt.figure(figsize = (7,4))
plt.bar (agg_id_income ['idcanalvenda'], agg_id_income['renda'], color = 'blue')
plt.ylabel ('Média de Idade')
plt.title ('Idade Média por Idade', loc ='center', fontsize = '18', fontweight = 'bold', color = 'black');
```

Representando a mudança de valores ao longo do tempo.

```python
salles_date = salles_right.groupby('Data').idcompra.nunique().reset_index()
salles_date
```

```python
fig = plt.figure(figsize = (20,12))

plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de vendas')
x = salles_date ['Data']
y = salles_date ['idcompra']

plt.plot(x,y, color = 'blue', linewidth = '2')
plt.show()
```

Visualizando a Média de preço de venda com frete

```python
agg_dep_price = round(salles_right.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dep_price
```

```python
fig = plt.figure(figsize = (34,26))
plt.bar (agg_dep_price ['Nome_Departamento'], agg_dep_price['Preço_com_frete'], color = 'blue')
plt.ylabel ('Preço Médio')
plt.title ('Preço Médio por Departamento', loc ='center', fontsize = '18', fontweight = 'bold', color = 'black');
plt.xticks(rotation = 60, ha='right');
```
